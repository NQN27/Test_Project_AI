```python
import folium
import pandas as pd
import numpy as np
import json
from folium import plugins

m = folium.Map(
    location=[16, 108],
    tiles='cartodbpositron',
    zoom_start=7,
)
from folium.plugins import Geocoder

Geocoder().add_to(m)
```

```python
df = pd.read_excel(r"C:\Users\FPTSHOP\Desktop\AI_project\A\Location.xlsx")
loc_matrix = np.array(df)
```

```python
#assume this path
path = ['HaNoi', 'BacNinh','HaiDuong', 'HaiPhong']
newpath = [["START",21.004355964700412, 105.84656236663436]] #add the start point
for name in path:
    for city in loc_matrix:
        if city[0] == name:
            coor = city[1].split(',')
            latitude = float(coor[0])
            longitude = float(coor[1].strip())
            newpath.append([name, latitude, longitude])
newpath.append(['ENDS',20.864056684118772, 106.71797164360206])
```

```python
for city in newpath:
    if city[0] == 'START':
        pop = 'your starting point'
        col = 'lightgreen'
    elif city[0] == 'ENDS':
        pop = 'your ending point'
        col = 'red'
    else:
        pop = 'in a route'
        col = 'orange'
    folium.Marker(
    location = city[1:],
    popup = pop,
    tooltip = city[0],
    icon = folium.Icon(color = col)
    ).add_to(m)
```

```python
for city in loc_matrix:
    if city[0] not in path:
        coor = city[1].split(',')
        latitude = float(coor[0])
        longitude = float(coor[1].strip())
        folium.Marker(
            location = [latitude, longitude],
            popup = 'not in any route',
            tooltip = city[0]
        ).add_to(m)
```

```python
lines = []
for i in newpath:
    lines.append(tuple(i[1:]))

folium.PolyLine(lines, color = 'red', weight = 5).add_to(m)

m
```
